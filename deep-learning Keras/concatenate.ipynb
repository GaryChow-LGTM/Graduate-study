```python
from keras.models import Model,Sequential
from keras.layers import Concatenate, Dense, LSTM, Input, concatenate
from keras.optimizers import Adagrad
from keras.utils import plot_model
from keras import layers

first_input = Input(shape=(2, ))
first_dense = Dense(1,activation='sigmoid' )(first_input)
"""
可以直接在Dense()中输入相关的参数，比如activation激活函数等。
"""

second_input = Input(shape=(2, ))
second_dense = Dense(1,activation='sigmoid' )(second_input)

third_input = Input(shape=(2, ))
third_dense = Dense(1,activation='sigmoid' )(third_input)

merge_one = concatenate([first_dense, second_dense, third_dense])
merge_two = concatenate([first_dense, second_dense, third_dense])

"merge_three = concatenate([merge_one, merge_two])"

model = Model(inputs=[first_input, second_input, third_input], outputs=[merge_one, merge_two])
ada_grad = Adagrad(lr=0.1, epsilon=1e-08, decay=0.0)
model.compile(optimizer=ada_grad, loss='binary_crossentropy',
               metrics=['accuracy'])
```

```python
plot_model(model,to_file='model.png')
```

```python
forth_input = Input(shape=(2, ))
forth_dense = Dense(1,activation='sigmoid' )(forth_input)

five_input = Input(shape=(2, ))
five_dense = Dense(1,activation='sigmoid' )(five_input)

six_input = Input(shape=(2, ))
six_dense = Dense(1,activation='sigmoid' )(six_input)

merge_forth = concatenate([forth_dense, five_dense, six_dense])
merge_five = concatenate([forth_dense, five_dense, six_dense])

"merge_six = concatenate([merge_forth, merge_five])"

model = Model(inputs=[forth_input, five_input, six_input], outputs=[merge_forth, merge_five])
ada_grad = Adagrad(lr=0.1, epsilon=1e-08, decay=0.0)
model.compile(optimizer=ada_grad, loss='binary_crossentropy',
               metrics=['accuracy'])
```

```python
plot_model(model,to_file='model2.png')
```

```python
merge_seven0 = concatenate([merge_one, merge_two, merge_forth, merge_five])
merge_seven1 = concatenate([merge_one, merge_two, merge_forth, merge_five])
merge_seven2 = concatenate([merge_one, merge_two, merge_forth, merge_five])
merge_seven3 = concatenate([merge_one, merge_two, merge_forth, merge_five])


model = Model(inputs=[first_input, second_input, third_input, forth_input, five_input, six_input], outputs=[merge_seven0, merge_seven1
                                                                                                           ,merge_seven2, merge_seven3])


ada_grad = Adagrad(lr=0.1, epsilon=1e-08, decay=0.0)
model.compile(optimizer=ada_grad, loss='binary_crossentropy',
               metrics=['accuracy'])
```

```python
plot_model(model,to_file='model3.png')
```
